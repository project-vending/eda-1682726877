```
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```
# Load the dataset from the given link
url = "https://raw.githubusercontent.com/midhunmohank/INFO6105/main/advertising.csv"
df = pd.read_csv(url)

```

```
# Check for null values in the dataset
print(df.isnull().sum())

```

```
# Print the head of the dataset
print(df.head())

```

```
# Get summary statistics of the dataset
print(df.describe())

```

```
# Visualize the distributions of the variables in the dataset
sns.pairplot(df, x_vars=["TV", "Radio", "Newspaper"], y_vars="Sales", height=7, aspect=0.7, kind="reg")
plt.show()

```

```
# Plot the correlation matrix of the variables in the dataset
sns.heatmap(df.corr(), annot=True)
plt.show()
```
